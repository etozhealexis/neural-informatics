```python
import numpy as np
from RoseblattPerceptron import RosenblattPerceptron
from BuildView import buildView
```

### 1 множество

```python
pInput1 = np.array([[2.6, -1.1], [3, 2.5], [-1.3, -2.9], [2.9, 4.4], [-1.8, 1.7], [-0.7, 3.3]])
tInput1 = np.array([1, 1, 1, 1, 0, 0])

perceptron1 = RosenblattPerceptron()
perceptron1.train(pInput1, tInput1, epochs = 2, showCoeffs = True)
```

```python
perceptron1.train(pInput1, tInput1, epochs = 50, showCoeffs = True)
```

```python
randomDotsToTest1 = perceptron1.testRandom(3)
buildView(perceptron1, pInput1, tInput1, randomDotsToTest1[0], randomDotsToTest1[1])
```

### 2 Множество
